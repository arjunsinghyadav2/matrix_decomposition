```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1,2,3],
     [2,1,3],
     [3,4,7]]
fig,ax = plt.subplots(figsize=(3,3))
ax.matshow(A, cmap='Blues')
for (i,j), val in np.ndenumerate(A):
    ax.text(j,i,f"{val}")
plt.show()
```

A contains column space and row space, a column space is unique columns of A that can not be created by taking combinations of other columns of A and similarly, row space has same idea for rows of A.

So if we consider C(A)<br/>
we have:

```python
#C(A)
C = np.array([[1,2,3], [2,1,4]])  # column space of A
R = np.array([[1,0,1],[0,1,1]])  # row space of A
#A = CR
fig, ax = plt.subplots(1, 3, figsize=(9, 3))

# Plot C
ax[0].matshow(C.T, cmap='Blues')
for (i, j), val in np.ndenumerate(C.T):
    ax[0].text(j, i, f"{val}", va='center', ha='center')
ax[0].set_title("C")

# Plot R
ax[1].matshow(R, cmap='Blues')
for (i, j), val in np.ndenumerate(R):
    ax[1].text(j, i, f"{val}", va='center', ha='center')
ax[1].set_title("R")

# Reconstruct A and plot
A_reconstructed = np.dot(C.T, R)
ax[2].matshow(A_reconstructed, cmap='Blues')
for (i, j), val in np.ndenumerate(A_reconstructed):
    ax[2].text(j, i, f"{val}", va='center', ha='center')
ax[2].set_title("A reconstructed from C and R")

plt.tight_layout()
plt.show()
```

```python
v = np.array([1,1,0])
w = np.array([0,1,1])
#if z is perpindicular to both v and w then it is in the null space of A
z = np.array([1,-1,1])  # z.v = 0 and z.w = 0
#plot the three vectors in 3D space
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='r', label='v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='g', label='w')
ax.quiver(0, 0, 0, z[0], z[1], z[2], color='b', label='z')
ax.set_xlim([ -1, 2])
ax.set_ylim([ -1, 2])
ax.set_zlim([ -1, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```

```python
#if u is not on the place vw then it is not a linear combination of v and w
u = np.array([2,-2,2])
print("u.v:", np.dot(u,v))
print("u.w:", np.dot(u,w))
print("u.z:", np.dot(u,z))
#plot the three vectors in 3D space
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='r', label='v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='g', label='w')
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='b', label='u')
ax.set_xlim([ -1, 2])
ax.set_ylim([ -1, 2])
ax.set_zlim([ -1, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```
